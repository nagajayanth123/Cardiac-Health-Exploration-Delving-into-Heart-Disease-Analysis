# Plot the Correlation Matrix

```python
plt.figure(figsize=(20,10))
corr_mtx = sns.heatmap(data.corr(),annot=True)
corr_mtx
```

```python
corr_val=data.corr()
target_correlation = corr_val['target']
significant_correlations = target_correlation[abs(target_correlation) > 0 ]  # Example threshold: 0.5
print("Significant Correlations with Target:")
print(significant_correlations.keys())
```

# #Find Gender Distribution According to The Target Variable

```python
sns.countplot(x='sex',hue='target',data=data)
plt.xticks([0,1],['Female','Male'])
plt.legend(labels=['No-Disease','Disease'])
plt.show()
```

# Check Age Distribution In The Dataset

# Check Chest Pain Type

```python
sns.countplot(data['cp'])
plt.xticks([0,1,2,3],['typicla-angina','atypical-angina','non-aginal-pain','asymptotic'])
plt.show()
```

# Show The Chest Pain Distribution As Per Target Variable

```python
sns.countplot(x='cp',hue='target',data=data)
plt.legend(labels=['No-Disease','Disease'])
plt.show()
```

# Show Fasting Blood Sugar Distribution According To Target Variable

```python
sns.countplot(x='fbs',hue='target',data=data)
plt.legend(labels=['No-Disease','Disease'])
plt.show()
```

# Compare Resting Blood Pressure As Per Sex Column

# Plot Continuous Variables

```python
cat_val=[]
cont_val=[]

for col in data.columns:
    if data[col].nunique()<=10:
        cat_val.append(col)
    else:
        cont_val.append(col)
```
